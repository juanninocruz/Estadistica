# Momento 2: Selección y análisis de información

## 2.1 Consulta y recopilación de información sobre análisis de ventas

De acuerdo con publicaciones especializadas en analítica de negocios y ciencia de datos, el análisis de ventas es una herramienta fundamental para evaluar el rendimiento comercial, identificar patrones de consumo y tomar decisiones informadas. Un análisis de ventas bien fundamentado no solo permite describir el comportamiento del negocio, sino también anticipar necesidades, optimizar recursos y mejorar la toma de decisiones (Harvard Business Review, 2020). Según HubSpot (2022), el análisis de ventas es el proceso mediante el cual una empresa examina datos sobre sus transacciones comerciales para entender qué productos se venden mejor, cuándo ocurren los mayores picos de venta y cómo se puede aumentar el rendimiento de las ventas. A partir de la revisión bibliográfica, se definieron los siguientes aspectos clave, que deben ser aplicados en el desarrollo de los análisis y proyectos:

- 📈 **Volumen de ventas:** (unidades vendidas por producto y por fecha) es fundamental para entender cuáles productos tienen mayor rotación, así como para estimar la demanda diaria y planificar el inventario. Este indicador permite conocer el comportamiento del consumidor y detectar productos de alta o baja salida (Tableau, s.f.).

- 💰 **Ingresos generados:** (Precio × Unidades vendidas) permite medir el impacto económico real de cada producto, más allá de su frecuencia de venta. Por ejemplo, un producto que se vende poco pero tiene alto precio puede generar más ingresos que otro con mayor rotación. Esto ayuda a identificar cuáles productos son más rentables (HubSpot, 2022).

- 🏆 **Identificación de productos destacados:** (más vendidos, menos vendidos, más rentables) permite establecer prioridades comerciales. Esta comparación facilita la toma de decisiones sobre qué productos promover, reabastecer o incluso reemplazar. Además, contribuye a evaluar la eficacia de estrategias de venta previas (Harvard Business Review, 2020).

- 📅 **Tendencias y estacionalidades:** (por día del mes o de la semana) es esencial para descubrir patrones y tendencias. Por ejemplo, es común que las ventas aumenten en fines de semana o fechas festivas como Navidad. Detectar estos comportamientos permite anticiparse a la demanda y planificar mejor el abastecimiento (Tableau, s.f.).

- 📊 **Visualización de datos:** El uso de gráficos (líneas, barras, tortas, etc.) facilita la comprensión de la información, tanto para el equipo de trabajo como para el dueño del negocio. La visualización ayuda a identificar rápidamente comportamientos inusuales, comparar productos y comunicar hallazgos de forma clara y efectiva (McKinney, 2017; VanderPlas, 2016).

- 🔢 **Preparación y agrupamiento de datos:** Antes de cualquier análisis, es clave aplicar una correcta limpieza y transformación de los datos: convertir fechas al formato adecuado, agrupar por variables como producto o fecha, y verificar la coherencia de precios y unidades. Según McKinney (2017), estas prácticas aseguran la calidad del análisis y evitan interpretaciones erróneas.

- 📉 **Uso de herramientas analíticas:** se recomienda el uso de herramientas como Excel, Python (pandas, matplotlib, seaborn) o Power BI, ya que facilitan tanto el procesamiento de los datos como la generación de dashboards e informes interactivos.


## 2.2 Métodos y herramientas de análisis de datos y visualización

Para realizar un análisis de ventas efectivo, es fundamental seleccionar herramientas que permitan procesar, transformar, analizar y visualizar los datos de manera estructurada, clara y accesible. Entre las cuales podemos encontrar:

- 🐍 **Pandas:** es una biblioteca de Python especializada en el manejo y análisis de estructuras de datos tabulares (como archivos .csv). Permite realizar tareas como carga de datos, limpieza, filtrado, agrupación por categorías (por ejemplo, por producto o fecha), y cálculo de métricas estadísticas como promedios, sumas o desviaciones estándar (McKinney, 2017).

- 🐍 **Matplotlib:** es una biblioteca de visualización en 2D que permite crear gráficos como líneas, barras, histogramas o tortas. Es altamente personalizable y facilita la representación gráfica de métricas clave, como ventas por día o comparaciones entre productos (VanderPlas, 2016).

- 🐍 **Seaborn:** está construida sobre Matplotlib, pero permite crear gráficos estadísticos con un estilo más estético y menos código. Es ideal para hacer comparaciones entre variables, visualizar tendencias o estudiar correlaciones en el comportamiento de las ventas.

- 🐍 **NumPy:** Permite trabajar con arreglos multidimensionales y realizar operaciones matemáticas complejas de forma eficiente, lo cual resulta útil para analizar series temporales, cálculos estadísticos y transformaciones numéricas de datos.

- 🐍 **SciPy:** complementa a NumPy al ofrecer funciones estadísticas y científicas más avanzadas, como pruebas de hipótesis, correlaciones, integraciones y optimización. Es ideal para análisis más profundos donde se requiere validar patrones mediante métodos estadísticos formales.

- 🐍 **Statsmodels:** biblioteca especializada en análisis estadístico, que permite aplicar modelos como regresiones lineales, análisis de varianza (ANOVA) y modelos de series temporales.

- 📊 **Power BI:** es una herramienta de visualización interactiva desarrollada por Microsoft, ampliamente utilizada en el mundo empresarial para transformar datos en dashboards dinámicos y fáciles de interpretar. Permite importar archivos .csv, generar gráficos interactivos (barras, líneas, tortas, mapas) y combinar múltiples métricas en un mismo informe (Tableau Software, s.f.).

- 📊 **Looker Studio:** es una plataforma de visualización gratuita que permite conectar datos desde múltiples fuentes (Google Sheets, BigQuery, etc.) y crear dashboards interactivos. Es ampliamente usada por equipos de marketing y análisis de negocios por su facilidad de uso e integración con el ecosistema de Google.

- 📊 **Tableau**: es una herramienta de visualización interactiva muy utilizada en entornos empresariales por su facilidad de uso y capacidad para crear dashboards dinámicos sin necesidad de programar (Tableau Software, s.f.).

```python
# prompt: Muestra la cantidad de productos vendidos de cada tipo

product_sales = data.groupby('Producto')['Unidades Vendidas'].sum().reset_index()
product_sales
```

```python
# prompt: muestra una grafica de ingresos respecto a producto de nombres deja la primera palabra de cada producto y los ingresos deja los en orden lo los miles con grid on y etiquetas de valores

import matplotlib.pyplot as plt
# Calcular ingresos por producto
ingresos_por_producto = data.groupby('Producto')['Ingresos'].sum().reset_index()

# Extraer la primera palabra del nombre del producto
ingresos_por_producto['Producto_Corto'] = ingresos_por_producto['Producto'].apply(lambda x: x.split()[0])

# Ordenar los ingresos de forma descendente
ingresos_por_producto = ingresos_por_producto.sort_values(by='Ingresos', ascending=False)

# Crear la gráfica
plt.figure(figsize=(10, 6))
bars = plt.bar(ingresos_por_producto['Producto_Corto'], ingresos_por_producto['Ingresos'], color='skyblue')

# Formatear los ingresos en miles
plt.gca().yaxis.set_major_formatter(plt.FuncFormatter(lambda x, loc: '{:,}'.format(int(x)).replace(',', '.') if x >= 1000 else int(x)))

# Añadir grid
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Añadir etiquetas de valor a cada barra
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2., yval, '{:,.0f}'.format(yval).replace(',', '.'), ha='center', va='bottom')

plt.xlabel('Producto')
plt.ylabel('Ingresos')
plt.title('Ingresos por Producto')
plt.show()
```

```python
# prompt: genera grafica de ingresos por cada vendedor

import matplotlib.pyplot as plt
# Calcular ingresos por vendedor
ingresos_por_vendedor = data.groupby('Vendedor')['Ingresos'].sum().reset_index()

# Ordenar los ingresos de forma descendente
ingresos_por_vendedor = ingresos_por_vendedor.sort_values(by='Ingresos', ascending=False)

# Crear la gráfica
plt.figure(figsize=(10, 6))
bars = plt.bar(ingresos_por_vendedor['Vendedor'], ingresos_por_vendedor['Ingresos'], color='lightgreen')

# Formatear los ingresos en miles
plt.gca().yaxis.set_major_formatter(plt.FuncFormatter(lambda x, loc: '{:,}'.format(int(x)).replace(',', '.') if x >= 1000 else int(x)))

# Añadir grid
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Añadir etiquetas de valor a cada barra
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2., yval, '{:,.0f}'.format(yval).replace(',', '.'), ha='center', va='bottom')

plt.xlabel('Vendedor')
plt.ylabel('Ingresos')
plt.title('Ingresos por Vendedor')
plt.show()
```

```python
# prompt: genera un pie chart utilizando la biblotec plotly que muestre la distribuccion de ingresos por vendedor y que de tooltip tenga el % del total, el ingreso por vendedor y su nombre

import plotly.express as px

# Calcular ingresos por vendedor
ingresos_por_vendedor = data.groupby('Vendedor')['Ingresos'].sum().reset_index()

# Crear el pie chart interactivo con Plotly
fig = px.pie(ingresos_por_vendedor,
             values='Ingresos',
             names='Vendedor',
             title='Distribución de Ingresos por Vendedor',
             hover_data=['Ingresos'], # Añadir ingresos al tooltip
             labels={'Ingresos':'Ingresos Totales'}) # Cambiar etiqueta en tooltip

# Actualizar el tooltip para mostrar el porcentaje, ingresos y nombre del vendedor
fig.update_traces(textposition='inside', textinfo='percent+label',
                  hovertemplate="<b>%{label}</b><br>Ingresos: %{value:,.0f}<br>Porcentaje: %{percent}")

fig.show()
```

```python
# prompt: genera unos indicadores con plotly que muestren el producto mas vendido, menos vendido, mas ingresos, menos ingresos

# Calcular el producto más vendido (en unidades)
producto_mas_vendido_unidades = product_sales.loc[product_sales['Unidades Vendidas'].idxmax()]

# Calcular el producto menos vendido (en unidades)
producto_menos_vendido_unidades = product_sales.loc[product_sales['Unidades Vendidas'].idxmin()]

# Calcular el producto con más ingresos
producto_mas_ingresos = ingresos_por_producto.loc[ingresos_por_producto['Ingresos'].idxmax()]

# Calcular el producto con menos ingresos
producto_menos_ingresos = ingresos_por_producto.loc[ingresos_por_producto['Ingresos'].idxmin()]


# Crear los indicadores con Plotly
import plotly.graph_objects as go

fig = go.Figure()

# Indicador: Producto Más Vendido (Unidades)
fig.add_trace(go.Indicator(
    mode="number",
    value=producto_mas_vendido_unidades['Unidades Vendidas'],
    title={"text": f"Producto Más Vendido<br><span style='font-size:0.8em;color:gray'>({producto_mas_vendido_unidades['Producto']})</span>"},
    domain={'x': [0, 0.25], 'y': [0.5, 1]}
))

# Indicador: Producto Menos Vendido (Unidades)
fig.add_trace(go.Indicator(
    mode="number",
    value=producto_menos_vendido_unidades['Unidades Vendidas'],
    title={"text": f"Producto Menos Vendido<br><span style='font-size:0.8em;color:gray'>({producto_menos_vendido_unidades['Producto']})</span>"},
    domain={'x': [0.35, 0.60], 'y': [0.5, 1]}
))

# Indicador: Producto con Más Ingresos
fig.add_trace(go.Indicator(
    mode="number",
    value=producto_mas_ingresos['Ingresos'],
    title={"text": f"Producto con Más Ingresos<br><span style='font-size:0.8em;color:gray'>({producto_mas_ingresos['Producto']})</span>"},
     number={'prefix': "$", 'valueformat': '.,0f'},
    domain={'x': [0, 0.25], 'y': [0, 0.45]}
))

# Indicador: Producto con Menos Ingresos
fig.add_trace(go.Indicator(
    mode="number",
    value=producto_menos_ingresos['Ingresos'],
    title={"text": f"Producto con Menos Ingresos<br><span style='font-size:0.8em;color:gray'>({producto_menos_ingresos['Producto']})</span>"},
     number={'prefix': "$", 'valueformat': '.,0f'},
    domain={'x': [0.35, 0.60], 'y': [0, 0.45]}
))


fig.update_layout(
    grid = {'rows': 2, 'columns': 2, 'pattern': "independent"},
    title="Indicadores Clave de Ventas por Producto"
)

fig.show()
```

```python
# prompt: genera unas cards  con plotly donde esten en ranking de ingresos los vendedores, que se tenga la opcion de drill down en estos vendedores para ver que productos estan vendiendo

# Calcular ingresos por vendedor
ingresos_por_vendedor = data.groupby('Vendedor')['Ingresos'].sum().reset_index()

# Ordenar los vendedores por ingresos de forma descendente
ingresos_por_vendedor = ingresos_por_vendedor.sort_values(by='Ingresos', ascending=False)

# Crear las cards interactivas para cada vendedor
for index, row in ingresos_por_vendedor.iterrows():
    vendedor = row['Vendedor']
    ingresos_totales_vendedor = row['Ingresos']

    # Filtrar los datos para el vendedor actual
    data_vendedor = data[data['Vendedor'] == vendedor]

    # Calcular los ingresos por producto para este vendedor
    ingresos_productos_vendedor = data_vendedor.groupby('Producto')['Ingresos'].sum().reset_index()
    ingresos_productos_vendedor = ingresos_productos_vendedor.sort_values(by='Ingresos', ascending=False)

    # Crear la figura de la card para el vendedor
    fig = go.Figure(go.Indicator(
        mode = "number+delta",
        value = ingresos_totales_vendedor,
        number = {'prefix': "$", 'valueformat': '.,0f'},
        delta = {'reference': ingresos_por_vendedor['Ingresos'].mean(), 'relative': True, 'valueformat': '.1%'}, # Comparación con el promedio
        title = {"text": f"Ingresos Totales: {vendedor}"},
        domain = {'x': [0, 1], 'y': [0.6, 1]}
    ))

    # Crear la tabla con el detalle de productos vendidos por el vendedor
    fig.add_trace(go.Table(
        header=dict(values=['Producto', 'Ingresos'],
                    fill_color='paleturquoise',
                    align='left'),
        cells=dict(values=[ingresos_productos_vendedor['Producto'],
                           # Corrected the format specifier to include thousands separator before the decimal point
                           ingresos_productos_vendedor['Ingresos'].apply(lambda x: f"${x:,.0f}")], # Formato de moneda
                   fill_color='lavender',
                   align='left'),
        domain = {'x': [0, 1], 'y': [0, 0.5]} # Posicionar la tabla debajo del indicador
    ))

    fig.update_layout(
        title_text=f'Performance de Ventas de {vendedor}',
        height=400 # Ajustar la altura de la card
    )

    fig.show()
```

```python
# prompt: genera un codigo que integre todo el trabajo con dash

# Load the data (assuming the data is already loaded and preprocessed as in the preceding code)
# Replace this with the actual loading logic if needed
# drive.mount('/content/drive')
# file_path = '/content/drive/My Drive/ventas2024_completo.csv'
# data = pd.read_csv(file_path, parse_dates=['Fecha'])

# Assuming 'data' DataFrame is already available from the preceding code
# Calculate aggregated data needed for the dashboard

# Calculate total units sold per product
product_sales = data.groupby('Producto')['Unidades Vendidas'].sum().reset_index()

# Calculate total revenue per product
ingresos_por_producto = data.groupby('Producto')['Ingresos'].sum().reset_index()
ingresos_por_producto['Producto_Corto'] = ingresos_por_producto['Producto'].apply(lambda x: x.split()[0])
ingresos_por_producto = ingresos_por_producto.sort_values(by='Ingresos', ascending=False)

# Calculate total revenue per seller
ingresos_por_vendedor = data.groupby('Vendedor')['Ingresos'].sum().reset_index()
ingresos_por_vendedor = ingresos_por_vendedor.sort_values(by='Ingresos', ascending=False)

# Calculate daily sales
daily_sales = data.groupby('Fecha')[['Unidades Vendidas', 'Ingresos']].sum().reset_index()

# Create the Dash app
app = dash.Dash(__name__, external_stylesheets=[dbc.themes.BOOTSTRAP])

# Define the layout of the app
app.layout = dbc.Container([
    html.H1("Análisis de Ventas - Tienda de Abarrotes (Diciembre 2024)", className="text-center my-4"),

    dbc.Row([
        dbc.Col(
            dcc.Graph(id='ingresos-por-producto'),
            width=6
        ),
        dbc.Col(
            dcc.Graph(id='ingresos-por-vendedor-pie'),
            width=6
        ),
    ]),

    dbc.Row([
        dbc.Col(
            dcc.Graph(id='daily-sales-line'),
            width=12
        ),
    ]),

    dbc.Row([
        dbc.Col(
            dcc.Graph(id='ingresos-por-vendedor-bar'),
            width=12
        )
    ]),

    # Separated the product indicators and seller analysis into different rows
    dbc.Row([
        dbc.Col(
            dcc.Graph(id='product-performance-indicators'),
            width=12 # Now takes full width
        ),
    ]),

    dbc.Row([
         dbc.Col(
            [
                html.H3("Análisis por Vendedor", className="text-center"),
                dcc.Dropdown(
                    id='seller-dropdown',
                    options=[{'label': seller, 'value': seller} for seller in ingresos_por_vendedor['Vendedor']],
                    value=ingresos_por_vendedor['Vendedor'].iloc[0], # Default value to the top seller
                    clearable=False,
                    className="mb-3"
                ),
                dcc.Graph(id='seller-product-distribution')
            ],
            width=12 # Now takes full width
        ),
    ]),


], fluid=True)

# Define callbacks to update graphs
@app.callback(
    Output('ingresos-por-producto', 'figure'),
    Input('ingresos-por-producto', 'id') # Dummy input
)
def update_ingresos_por_producto_graph(_):
    fig = px.bar(ingresos_por_producto,
                 x='Producto_Corto',
                 y='Ingresos',
                 title='Ingresos por Producto',
                 labels={'Producto_Corto': 'Producto', 'Ingresos': 'Ingresos'},
                 template='plotly_white')
    fig.update_layout(xaxis_tickangle=-45)
    fig.update_traces(texttemplate='$%{y:,.0f}', textposition='outside')
    return fig

@app.callback(
    Output('ingresos-por-vendedor-pie', 'figure'),
    Input('ingresos-por-vendedor-pie', 'id') # Dummy input
)
def update_ingresos_por_vendedor_pie(_):
    fig = px.pie(ingresos_por_vendedor,
                 values='Ingresos',
                 names='Vendedor',
                 title='Distribución de Ingresos por Vendedor',
                 hover_data=['Ingresos'],
                 labels={'Ingresos':'Ingresos Totales'},
                 template='plotly_white')
    fig.update_traces(textposition='inside', textinfo='percent+label',
                      hovertemplate="<b>%{label}</b><br>Ingresos: %{value:,.0f}<br>Porcentaje: %{percent}")
    return fig

@app.callback(
    Output('daily-sales-line', 'figure'),
    Input('daily-sales-line', 'id') # Dummy input
)
def update_daily_sales_line(_):
    fig = go.Figure()
    fig.add_trace(go.Scatter(x=daily_sales['Fecha'], y=daily_sales['Ingresos'], mode='lines+markers', name='Ingresos Diarios'))
    fig.add_trace(go.Scatter(x=daily_sales['Fecha'], y=daily_sales['Unidades Vendidas'], mode='lines+markers', name='Unidades Vendidas Diarias', yaxis='y2'))

    fig.update_layout(
        title='Ingresos y Unidades Vendidas Diarias',
        xaxis_title='Fecha',
        yaxis_title='Ingresos',
        yaxis2=dict(
            title='Unidades Vendidas',
            overlaying='y',
            side='right'
        ),
        template='plotly_white'
    )
    return fig


@app.callback(
    Output('ingresos-por-vendedor-bar', 'figure'), # Output for the new bar chart
    Input('ingresos-por-vendedor-bar', 'id') # Dummy input
)
def update_ingresos_por_vendedor_bar(_):
    fig = px.bar(ingresos_por_vendedor,
                 x='Vendedor',
                 y='Ingresos',
                 title='Ingresos Totales por Vendedor',
                 labels={'Vendedor': 'Vendedor', 'Ingresos': 'Ingresos'},
                 template='plotly_white')
    fig.update_layout(xaxis_tickangle=-45)
    fig.update_traces(texttemplate='$%{y:,.0f}', textposition='outside')
    return fig


@app.callback(
    Output('product-performance-indicators', 'figure'),
    Input('product-performance-indicators', 'id') # Dummy input
)
def update_product_indicators(_):
    # Calculate the most and least sold product (by units)
    producto_mas_vendido_unidades = product_sales.loc[product_sales['Unidades Vendidas'].idxmax()]
    producto_menos_vendido_unidades = product_sales.loc[product_sales['Unidades Vendidas'].idxmin()]

    # Calculate the product with the most and least revenue
    producto_mas_ingresos = ingresos_por_producto.loc[ingresos_por_producto['Ingresos'].idxmax()]
    producto_menos_ingresos = ingresos_por_producto.loc[ingresos_por_producto['Ingresos'].idxmin()]

    fig = go.Figure()

    fig.add_trace(go.Indicator(
        mode="number",
        value=producto_mas_vendido_unidades['Unidades Vendidas'],
        title={"text": f"Producto Más Vendido<br><span style='font-size:0.8em;color:gray'>({producto_mas_vendido_unidades['Producto']})</span>"},
        domain={'x': [0, 0.5], 'y': [0.5, 1]}
    ))

    fig.add_trace(go.Indicator(
        mode="number",
        value=producto_menos_vendido_unidades['Unidades Vendidas'],
        title={"text": f"Producto Menos Vendido<br><span style='font-size:0.8em;color:gray'>({producto_menos_vendido_unidades['Producto']})</span>"},
        domain={'x': [0.5, 1], 'y': [0.5, 1]}
    ))

    fig.add_trace(go.Indicator(
        mode="number",
        value=producto_mas_ingresos['Ingresos'],
        title={"text": f"Producto con Más Ingresos<br><span style='font-size:0.8em;color:gray'>({producto_mas_ingresos['Producto']})</span>"},
         number={'prefix': "$", 'valueformat': '.,0f'},
        domain={'x': [0, 0.5], 'y': [0, 0.45]}
    ))

    fig.add_trace(go.Indicator(
        mode="number",
        value=producto_menos_ingresos['Ingresos'],
        title={"text": f"Producto con Menos Ingresos<br><span style='font-size:0.8em;color:gray'>({producto_menos_ingresos['Producto']})</span>"},
         number={'prefix': "$", 'valueformat': '.,0f'},
        domain={'x': [0.5, 1], 'y': [0, 0.45]}
    ))

    fig.update_layout(
        grid = {'rows': 2, 'columns': 2, 'pattern': "independent"},
        title="Indicadores Clave de Ventas por Producto",
        template='plotly_white'
    )
    return fig


@app.callback(
    Output('seller-product-distribution', 'figure'),
    Input('seller-dropdown', 'value')
)
def update_seller_product_distribution(selected_seller):
    filtered_data = data[data['Vendedor'] == selected_seller]
    ingresos_productos_vendedor = filtered_data.groupby('Producto')['Ingresos'].sum().reset_index()
    ingresos_productos_vendedor = ingresos_productos_vendedor.sort_values(by='Ingresos', ascending=False)

    fig = px.bar(ingresos_productos_vendedor,
                 x='Producto',
                 y='Ingresos',
                 title=f'Distribución de Ingresos por Producto para {selected_seller}',
                 labels={'Producto': 'Producto', 'Ingresos': 'Ingresos'},
                 template='plotly_white')
    fig.update_layout(xaxis_tickangle=-45)
    fig.update_traces(texttemplate='$%{y:,.0f}', textposition='outside')
    return fig


# Run the app
if __name__ == '__main__':
    app.run(debug=True, mode='inline') # Use mode='inline' for Colab
```

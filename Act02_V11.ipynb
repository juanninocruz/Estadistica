```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import ttest_ind

# Assuming 'data2' DataFrame is already defined as in the provided code.
# Ensure 'age' is numeric
data2['age'] = pd.to_numeric(data2['age'], errors='coerce')

# Perform t-test for 'age' vs. 'stroke'
# Separate 'age' values by 'stroke' status
age_stroke = data2[data2['stroke'] == 1]['age'].dropna() # Drop NaNs for t-test
age_no_stroke = data2[data2['stroke'] == 0]['age'].dropna() # Drop NaNs for t-test

# Perform the t-test
t_statistic, p_value = ttest_ind(age_stroke, age_no_stroke)

print(f"T-statistic (Age vs Stroke): {t_statistic}")
print(f"P-value (Age vs Stroke): {p_value}")

# Perform t-test for 'avg_glucose_level' vs. 'stroke'
# Separate 'avg_glucose_level' values by 'stroke' status
glucose_stroke = data2[data2['stroke'] == 1]['avg_glucose_level'].dropna()
glucose_no_stroke = data2[data2['stroke'] == 0]['avg_glucose_level'].dropna()

# Perform the t-test
t_statistic, p_value = ttest_ind(glucose_stroke, glucose_no_stroke)

print(f"T-statistic (Avg Glucose Level vs Stroke): {t_statistic}")
print(f"P-value (Avg Glucose Level vs Stroke): {p_value}")

# Perform t-test for 'bmi' vs. 'stroke'
# Separate 'bmi' values by 'stroke' status
bmi_stroke = data2[data2['stroke'] == 1]['bmi'].dropna()
bmi_no_stroke = data2[data2['stroke'] == 0]['bmi'].dropna()

# Perform the t-test
t_statistic, p_value = ttest_ind(bmi_stroke, bmi_no_stroke)

print(f"T-statistic (BMI vs Stroke): {t_statistic}")
print(f"P-value (BMI vs Stroke): {p_value}")

# Assuming 'data2' DataFrame is already defined and preprocessed
# Map 'stroke' column to numeric if it's not already (ensure it's 0 and 1)
data2['stroke_numeric'] = data2['stroke'].map({0: 0, 1: 1})


# Separate numerical variables for those with stroke (1) and without stroke (0)
age_stroke = data2[data2['stroke_numeric'] == 1]['age']
age_no_stroke = data2[data2['stroke_numeric'] == 0]['age']

glucose_stroke = data2[data2['stroke_numeric'] == 1]['avg_glucose_level']
glucose_no_stroke = data2[data2['stroke_numeric'] == 0]['avg_glucose_level']

bmi_stroke = data2[data2['stroke_numeric'] == 1]['bmi']
bmi_no_stroke = data2[data2['stroke_numeric'] == 0]['bmi']

# Perform t-tests, handling potential NaNs by dropping them for the test
t_stat_age, p_val_age = ttest_ind(age_stroke.dropna(), age_no_stroke.dropna())
t_stat_glucose, p_val_glucose = ttest_ind(glucose_stroke.dropna(), glucose_no_stroke.dropna())
t_stat_bmi, p_val_bmi = ttest_ind(bmi_stroke.dropna(), bmi_no_stroke.dropna())

print(f"T-test results for Age vs Stroke: t-statistic = {t_stat_age:.4f}, p-value = {p_val_age:.4f}")
print(f"T-test results for Avg Glucose Level vs Stroke: t-statistic = {t_stat_glucose:.4f}, p-value = {p_val_glucose:.4f}")
print(f"T-test results for BMI vs Stroke: t-statistic = {t_stat_bmi:.4f}, p-value = {p_val_bmi:.4f}")


# Boxplots for visualization
plt.figure(figsize=(18, 6))

plt.subplot(1, 3, 1)
sns.boxplot(x='stroke_numeric', y='age', data=data2)
plt.title('Gráfica 43. t-student entre Edad y Accidente Cerebrovascular')
plt.xlabel('Accidente Cerebrovascular (0: No, 1: Sí)')
plt.ylabel('Edad')

plt.subplot(1, 3, 2)
sns.boxplot(x='stroke_numeric', y='avg_glucose_level', data=data2)
plt.title('Gráfica 44. t-student entre Nivel de Glucosa y Accidente Cerebrovascular')
plt.xlabel('Accidente Cerebrovascular (0: No, 1: Sí)')
plt.ylabel('Nivel de Glucosa Promedio')

plt.subplot(1, 3, 3)
sns.boxplot(x='stroke_numeric', y='bmi', data=data2)
plt.title('Gráfica 45. t-student entre IMC y Accidente Cerebrovascular')
plt.xlabel('Accidente Cerebrovascular (0: No, 1: Sí)')
plt.ylabel('IMC')

plt.tight_layout()
plt.show()

# Assuming 'data2' DataFrame is already defined and preprocessed
# Ensure 'hypertension' and 'heart_disease' are treated as categorical or convert them for t-test
# Note: t-test is typically for comparing the means of two *numerical* groups.
# You would usually use chi-squared for binary vs binary (like hypertension vs stroke).
# However, if you want to compare the *mean* of a *numerical* variable between the two groups defined by hypertension/heart_disease, you can do a t-test.
# Let's assume the goal is to see if the average 'age', 'avg_glucose_level', or 'bmi' is different between people with/without hypertension or heart disease.

# Convert binary variables to numeric if they are not already (assuming 0 and 1)
data2['hypertension_numeric'] = data2['hypertension'].astype(int)
data2['heart_disease_numeric'] = data2['heart_disease'].astype(int)


# T-test for numerical variables vs Hypertension
print("\n--- T-tests vs Hypertension ---")
for num_var in ['age', 'avg_glucose_level', 'bmi']:
    group_with = data2[data2['hypertension_numeric'] == 1][num_var].dropna()
    group_without = data2[data2['hypertension_numeric'] == 0][num_var].dropna()

    if len(group_with) > 1 and len(group_without) > 1: # Ensure enough samples in each group
        t_stat, p_val = ttest_ind(group_with, group_without)
        print(f"T-test results for {num_var} vs Hypertension: t-statistic = {t_stat:.4f}, p-value = {p_val:.4f}")
    else:
         print(f"Insufficient data to perform T-test for {num_var} vs Hypertension")


# T-test for numerical variables vs Heart Disease
print("\n--- T-tests vs Heart Disease ---")
for num_var in ['age', 'avg_glucose_level', 'bmi']:
    group_with = data2[data2['heart_disease_numeric'] == 1][num_var].dropna()
    group_without = data2[data2['heart_disease_numeric'] == 0][num_var].dropna()

    if len(group_with) > 1 and len(group_without) > 1: # Ensure enough samples in each group
        t_stat, p_val = ttest_ind(group_with, group_without)
        print(f"T-test results for {num_var} vs Heart Disease: t-statistic = {t_stat:.4f}, p-value = {p_val:.4f}")
    else:
        print(f"Insufficient data to perform T-test for {num_var} vs Heart Disease")

# Boxplots for numerical variables vs Hypertension
plt.figure(figsize=(18, 6))

plt.subplot(1, 3, 1)
sns.boxplot(x='hypertension_numeric', y='age', data=data2)
plt.title('Gráfica 46. t-student entre Hipertensión y Edad')
plt.xlabel('Hipertensión (0: No, 1: Sí)')
plt.ylabel('Edad')

plt.subplot(1, 3, 2)
sns.boxplot(x='hypertension_numeric', y='avg_glucose_level', data=data2)
plt.title('Gráfica 47. t-student entre Hipertensión y Nivel de Glucosa')
plt.xlabel('Hipertensión (0: No, 1: Sí)')
plt.ylabel('Nivel de Glucosa Promedio')

plt.subplot(1, 3, 3)
sns.boxplot(x='hypertension_numeric', y='bmi', data=data2)
plt.title('Gráfica 48. t-student entre Hipertensión e IMC')
plt.xlabel('Hipertensión (0: No, 1: Sí)')
plt.ylabel('IMC')

plt.tight_layout()
plt.show()

# Boxplots for numerical variables vs Heart Disease
plt.figure(figsize=(18, 6))

plt.subplot(1, 3, 1)
sns.boxplot(x='heart_disease_numeric', y='age', data=data2)
plt.title('Gráfica 49. t-student entre Enfermedad Cardiovascular y Edad')
plt.xlabel('Enfermedad Cardiovascular (0: No, 1: Sí)')
plt.ylabel('Edad')

plt.subplot(1, 3, 2)
sns.boxplot(x='heart_disease_numeric', y='avg_glucose_level', data=data2)
plt.title('Gráfica 50. t-student entre Enfermedad Cardiovascular y Nivel de Glucosa')
plt.xlabel('Enfermedad Cardiovascular (0: No, 1: Sí)')
plt.ylabel('Nivel de Glucosa Promedio')

plt.subplot(1, 3, 3)
sns.boxplot(x='heart_disease_numeric', y='bmi', data=data2)
plt.title('Gráfica 51. t-student entre Enfermedad Cardiovascular e IMC')
plt.xlabel('Enfermedad Cardiovascular (0: No, 1: Sí)')
plt.ylabel('IMC')

plt.tight_layout()
plt.show()

```
